```python
dataset_path='Wearable_Dataset/'
```

```python
states=os.listdir(dataset_path)
```

```python
signal_data={}
time_data={}
fs_dict={}
participants={}

for state in states:
    folder_path = f'{dataset_path}/{state}' 
    participants[state]=os.listdir(folder_path)
    signal_data[state], time_data[state], fs_dict[state] = RawData.read_signals(folder_path) # Returns three dictionaries with subjects info: raw signals (signal_data), temporal data ready to graph (time_data) and sample frequency for escha signal(fs_dict).
```

```python
df_stress = pd.DataFrame(signal_data['STRESS']).T
df_stress.columns = ['ACC', 'BVP', 'EDA', 'HR', 'tags', 'TEMP']
df_stress.head()
```

```python
df_time = pd.DataFrame(time_data['STRESS']).T
df_time.head()
```

```python
path = 'meta-data/'
df_stress.to_pickle(path + 'stress_signals_D2.pkl')
df_time.to_pickle(path + 'stress_time_D2.pkl')
```
